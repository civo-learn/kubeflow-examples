```python
from openai import OpenAI


client = OpenAI(base_url = 'http://localhost:8000/v1', api_key="test")
response = client.chat.completions.create(
  model="llama-2",
  messages=[
    {
      "role": "system",
      "content": "Hey, you are a smart LLM called Llama 2."
    },
    {
      "role": "user",
      "content": "I went to the market and bought 10 apples. I gave 2 apples to the neighbor and 2 to the repairman. I then went and bought 5 more apples and ate 1. How many apples did I remain with?\n\nLet's think step by step."
    }
  ],
  temperature=1,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
print(response)
```

```python
import requests
import json

text_to_translate = "Hello!"
url = "http://localhost:8000/v1/completions"

payload = {
    "prompt": f"Translate the following English text to French: \"{text_to_translate}\""
}

# Making the POST request
response = requests.post(url, json=payload)

if response.status_code == 200:
    print("Translation request was successful!")
    print("Response:", response.text)
else:
    print("Failed to get a successful response.")
    print("Status code:", response.status_code)
    print("Response:", response.text)
```
